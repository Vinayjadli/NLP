```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
from nltk.stem import WordNetLemmatizer
wl=WordNetLemmatizer()
```

```python
messages= X['title'].copy()
```

```python
main=[]
long=[]
for s in range(len(messages)):
    review= re.sub('[^a-zA-Z]',' ',messages[s])
    review= review.lower()
    review= review.split()
    main.append(review)
for word1 in main:
    l=[]
    for word in word1:
        if word not in stopwords.words('english'):
            l.append(ps.stem(word))
    long.append(l)
# now this is the longer method for removing the stopwords    
```

```python
corpus=[]
ans=[]
for s in range(len(messages)):
    review= re.sub('[^a-zA-Z]',' ',messages[s])
    review= review.lower()
    review= review.split()
    
    corpus=[ps.stem(word)for word in review if word not in stopwords.words('english')]
    corpus = ' '.join(corpus)
    ans.append(corpus)
# this is the short method for removing the stopwords
```

```python
ans
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size=10000
```

```python
#one_hot_rep = [one_hot(word,voc_size) for word in ans]
onehot_repr=[one_hot(words,voc_size)for words in ans] 
onehot_repr
```

### Embedding process

```python
# but for make every sentence of same length we have to find the highest length
max_length=0
for word in ans:
    word=word.split()
    if len(word)>max_length:
        max_length=len(word)
print(max_length)

# so from here we can see that the max length is 47
```

```python
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=max_length)
embedded_docs
```

```python
dimension=40
model=Sequential()
model.add(Embedding(voc_size,dimension,input_length=max_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
X_final=np.array(embedded_docs)
Y_final=np.array(Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X_final,Y_final,train_size=0.7,random_state=42)
```

#### Model Training

```python
model.fit(X_train,Y_train,validation_data=(X_test,Y_test),epochs=10,batch_size=64)
```
